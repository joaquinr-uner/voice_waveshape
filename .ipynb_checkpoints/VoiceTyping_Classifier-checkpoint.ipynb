# Library Imports


```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 29 15:53:57 2022

@author: joaquinruiz
"""
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split, StratifiedKFold, KFold
from sklearn.metrics import confusion_matrix, accuracy_score, balanced_accuracy_score, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from itertools import product, compress
import os
import warnings
```

#Function to construct average confusion matrices

```python
def plot_mean_confumat(cm_avr, cm_sd, ax, title, clas_lab, set_cbar={'cmap':'viridis','cbar_shrink':1,'cmax':None,'cmin':None}, fontsz = 12):
  if set_cbar['cmin'] or set_cbar['cmax']:
    cmap_min, cmap_max = set_cbar['cmin'], set_cbar['cmax']
  else:
    cmap_min, cmap_max = cm_avr.min(), cm_avr.max()

  im=ax.imshow(cm_avr,vmin=cmap_min, vmax=cmap_max,interpolation='nearest',cmap=set_cbar['cmap'])


  cc_min, cc_max = im.cmap(0), im.cmap(256)
  cbar = plt.colorbar(im,ax=ax,shrink=set_cbar['cbar_shrink'])
  cbar.ax.tick_params(labelsize=fontsz)

  thresh = (cmap_max + cmap_min) / 2.0
  n_classes = cm_avr.shape[0]
  for r,c in product(np.arange(n_classes),np.arange(n_classes)):
      color = cc_max if cm_avr[r, c] < thresh else cc_min
      text = ax.text(c,r,'{mean:.2f}\n ({std:.2f})'.format(mean=cm_avr[r,c],
                      std=cm_sd[r,c]),ha="center",va="center",color=color, fontsize = fontsz)
  ax.set(xticks=np.arange(n_classes), yticks=np.arange(n_classes),
         xticklabels=clas_lab, yticklabels=clas_lab,
         ylabel="True label", xlabel="Predicted label")
  for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(fontsz)
  ax.set_title(title,{'fontsize':fontsz})
  plt.tight_layout()
```

# Training-Validation with full database (Meei+Saar)

##Data Loading

```python
!git clone https://github.com/joaquinr-uner/voice_waveshape.git
```

```python
DB = 'All'
R = 18
N = 15
df = pd.read_csv('voice_waveshape/Features_VoiceTyping.csv')
X_org = df.iloc[:,:-2]
n_splits = 10
db = df['Database']
Y3_org = df['Type']
Y2 = np.zeros((X_org.shape[0],))
b = Y3_org==3
Y2[b] = 1

do_plot = False
```

```python
if DB != 'All':
  X_org = X_org[db==DB.lower()]
  X_org.index = [np.arange(0,X_org.shape[0])]
  Y3_org = Y3_org[db==DB.lower()]
  Y3_org.index = [np.arange(0,X_org.shape[0])]
  Y2 = np.zeros((X_org.shape[0],))
  b = Y3_org==3
  Y2[b] = 1
X_org
```

```python
Xnstd = X_org.iloc[:,np.concatenate(([0],np.arange(3,38)))]
Xnstd.columns
```

## Cascade SVM (Linear SVM + RB SVM)
Hiperparámetros a optimizar:


*   $C_{Lin}$. Rango: $1^{-3} - 10$. $L = 7$
*   $C_{RBF}$. Rango: $1^{-2} - 1000$. $L = 7$
*   $\gamma$. Rango: $\frac{0.01}{N_{features}} - \frac{10}{N_{features}}$. $L = 7$



```python
VCLin = np.logspace(-2,2,num=7)
VCRBF = np.logspace(-1,3,num=7)
Y = Y2
Y3 = Y3_org.to_numpy()

scaler = StandardScaler().fit(Xnstd)
X = scaler.transform(Xnstd)

Vgamm = 1/X.shape[1]*np.logspace(-2,2,num=7)
```

```python
VCLin, VCRBF, Vgamm
```

```python
it = product(VCLin,VCRBF,Vgamm)

V_scores = []
for C1, C2, gam in it:
  net1 = SVC(C = C1, kernel = 'linear', class_weight='balanced')
  net2 = SVC(C = C2, gamma = gam, class_weight='balanced')
  #net_svm = SVC()
  cv = StratifiedKFold(n_splits=n_splits, shuffle=True)
  #cv = KFold(n_splits=n_splits)
  score2 = np.zeros((n_splits))
  score3 = np.zeros((n_splits))
  acc_svm = np.zeros((n_splits))
  cm2 = np.zeros((n_splits,2,2))
  cm3 = np.zeros((n_splits,3,3))
  for i, (Train, Test) in enumerate(cv.split(X,Y)):
      Xtrain, Ytrain, Y3train, Xtest, Ytest, Y3test = X[Train], Y[Train], Y3[Train], X[Test], Y[Test], Y3[Test]
      net1.fit(Xtrain,Ytrain)
      y_pred = net1.predict(Xtest)
      X2train = Xtrain[Ytrain==0]
      Y2train = Y3train[Ytrain==0]
      tp = np.logical_and(y_pred==0,Ytest==0)
      fp = np.logical_and(y_pred==0,Ytest==1)
      indx_p = np.array(list(compress(range(len(tp+fp)),tp+fp)))
      X2test = Xtest[tp+fp]
      Y2test = Y3test[tp+fp]
      net2.fit(X2train,Y2train)
      y2_pred = net2.predict(X2test)
      y3_pred = 3*np.ones((Test.shape))
      y3_pred[indx_p] = y2_pred
      score2[i] = balanced_accuracy_score(Ytest, y_pred)
      score3[i] = balanced_accuracy_score(Y3test, y3_pred)
      cm2[i,:,:] = confusion_matrix(Ytest,y_pred,normalize='true')
      cm3[i,:,:] = confusion_matrix(Y3test,y3_pred,normalize='true')
  V_scores.append({'CLin':C1,'C_RBF':C2,'gamma':gam,'scores_bin':score2,'scores_multi':score3,'mean_score':np.mean(score3),'confumat2':cm2,'confumat3':cm3})
```

```python
mean_scores = [score['mean_score'] for score in V_scores]
np.argmax(mean_scores)
```

```python
best = V_scores[np.argmax(mean_scores)]
best
```

## Plotting

```python
fig, ax = plt.subplots(1,2,figsize=(21,21))
cbars = 0.3
fntsz = 24

cm2_avr = np.mean(best['confumat2'],axis=0)
cm2_sd = np.std(best['confumat2'],axis=0)

cm3_avr = np.mean(best['confumat3'],axis=0)
cm3_sd = np.std(best['confumat3'],axis=0)

title2 = f'SVM Lin. DB: {DB}' + '\n' + f'Average Balanced Accuracy (Std in brackets): {round(np.mean(best["scores_bin"]),2)} ({round(np.std(best["scores_bin"]),2)})'
plot_mean_confumat(cm2_avr, cm2_sd, ax[0], title2, ['1+2','3'], set_cbar = {'cmap':'pink','cbar_shrink':cbars, 'cmin':None, 'cmax':None}, fontsz = fntsz)

title3 = f'SVM Lin+RBF. DB: {DB}' + '\n' + f'Average Balanced Accuracy (Std in brackets): {round(np.mean(best["scores_multi"]),2)} ({round(np.std(best["scores_multi"]),2)})'
plot_mean_confumat(cm3_avr, cm3_sd, ax[1], title3, ['1','2','3'],set_cbar = {'cmap':'pink','cbar_shrink':cbars, 'cmin':None, 'cmax':None}, fontsz = fntsz)
```
